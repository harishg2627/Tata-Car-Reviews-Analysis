# Car Reviews

```python
#How many reviews are available for each Tata car model?
review_counts = df['Car_Name'].value_counts()
print(review_counts)

plt.figure(figsize=(10,5))
sns.barplot(x=review_counts.index, y=review_counts.values)
plt.xticks(rotation=45)
plt.title("Number of Reviews per Tata Car Model")
plt.ylabel("Count")
plt.xlabel("Car Model")
plt.show()
```

# Economy and Performance

```python
#Average economy and performance comparison per car
avg_data = df.groupby('Car_Name')[['economy', 'performance']].mean().reset_index()
avg_data.plot(x='Car_Name', kind='bar', figsize=(10,5))
plt.title("Average Economy and Performance by Car Model")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```

# Features and Ratings

```python
#Which features (like comfort, performance, looks) most affect ratings?
numeric_cols = df.select_dtypes(include=np.number).columns
corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Features and Rating")
plt.show()
```

# Positive and Negative Reviews Based On Models

```python
#Which car models have the most positive and negative reviews?
positive = df[df['overall'] >= 4]['Car_Name'].value_counts().head(5)
negative = df[df['overall'] <= 2]['Car_Name'].value_counts().head(5)
fig, ax = plt.subplots(1,2, figsize=(12,6))
ax[0].pie(positive.values, labels=positive.index, autopct='%1.1f%%')
ax[1].pie(negative.values, labels=negative.index, autopct='%1.1f%%')
ax[0].set_title("Positive Reviews")
ax[1].set_title("Negative Reviews")

plt.show()

```

# Customer Ratings

```python
#How do customers rate overall performance of cars?
plt.figure(figsize=(8,5))
sns.histplot(df['performance'], bins=10, kde=True)
plt.title("Distribution of Performance Ratings")
plt.xlabel("Performance Rating")
plt.show()
```

# Comfort Ratings

```python
#Compare comfort ratings across car models
plt.figure(figsize=(10,6))
sns.pointplot(x='Car_Name', y='comfort', data=df)
plt.xticks(rotation=45)
plt.title("Average Comfort Rating by Car Model")
plt.show()

```

# Most Liked Car Model

```python
#Which car models have the most liked reviews?
top_liked = df.groupby('Car_Name')['total_likes'].sum().sort_values(ascending=False)
top_liked.plot(kind='bar', figsize=(10,5))
plt.title("Total Likes by Car Model")
plt.ylabel("Sum of Likes")
plt.show()
```

# Value For Money Rating

```python
#Average Value for Money rating by each car model
avg_value = df.groupby('Car_Name')['value_for_money'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
plt.plot(avg_value.index, avg_value.values, marker='o')
plt.xticks(rotation=45)
plt.title("Average Value for Money by Car Model")
plt.ylabel("Average Score")
plt.tight_layout()
plt.show()

```

# Purchase Condition Distribution

```python
#How many users bought cars under different purchase conditions?
order = df['purchase_condition'].value_counts().index

sns.countplot(
    x='purchase_condition',
    hue='purchase_condition',
    data=df,
    order=order,
    palette='viridis',
    legend=False
)

plt.title("Purchase Condition Distribution (Sorted)")
plt.xlabel("Condition")
plt.ylabel("Number of Buyers")
plt.show()
```

# Comfort and Overall Ratings

```python
#Relationship between comfort and overall ratings
sns.lmplot(
    x='comfort',
    y='overall',
    data=df,
    scatter_kws={'s': 60},
    line_kws={'linewidth': 2},
    height=5
)
plt.title("Comfort vs Overall Rating")
plt.show()
```
